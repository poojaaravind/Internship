program 5(iii)

Top 10 ODI bowlers along with the records of their team and rating.

```python
page5_3=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling")
page5_3
soup5_3=BeautifulSoup(page5_3.content)
```

```python
bowlers=[]
records=[]
rating=[]
bowl=soup5_3.find_all('td',class_="table-body__cell rankings-table__name name")
for i in bowl:
    bowlers.append(i.text.replace('\n',''))
    
record=soup5_3.find_all('span',class_="table-body__logo-text")
for i in record:
    records.append(i.text.replace('\n',''))
    
rate=soup5_3.find_all('td',class_="table-body__cell rating")
for i in rate:
    rating.append(i.text.replace('\n',''))
```

```python
import pandas as pd
odibowlers=pd.DataFrame({})
odibowlers['Bowlers']=bowlers
odibowlers['Nation']=records
odibowlers['Rating']=rating
```

```python
Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.
```

```python
page6=requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
page6
soup6=BeautifulSoup(page6.content)
```

```python
teams=[]
matches=[]
points=[]
rating=[]
team=soup6.find_all('span',class_="u-hide-phablet")
for i in team:
    teams.append(i.text.replace('\n',''))
    
match=soup6.find_all('td',class_="table-body__cell u-center-text")
for i in match:
    matches.append(i.text.replace('\n',''))
    
point=soup6.find_all('td',class_="table-body__cell u-center-t")
for i in point:
    points.append(i.text.replace('\n',''))
    
rate=soup6.find_all('td',class_="table-body__cell u-text-right rating")
for i in rate:
    rating.append(i.text.replace('\n',''))
```

```python
len(matches)
```

```python
import pandas as pd
odiwomen=pd.DataFrame({})
odiwomen['Rating']=teams
odiwomen['Bowlers']=matches
odiwomen['Nation']=points
odiwomen['Rating']=rating
```

Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The
scraped data should include Product Name, Price, Image URL and Average Rating.

```python
page7=requests.get("https://www.amazon.in/s?k=smartphones+under+20000&ref=nb_sb_noss_2")
page7
soup7=BeautifulSoup(page7.content)
```

```python
product_name=[]
price=[]
image_url=[]
rating=[]
product=soup7.find_all('span',class_="a-size-medium a-color-base a-text-normal")
for i in product:
    product_name.append(i.text.replace('\n',''))
    
prices=soup7.find_all('span',class_="a-price-whole")
for i in prices:
    price.append(i.text.replace('\n',''))
    
image=soup7.find_all('img',class_="s-image")
for i in image:
    image_url.append(i)
    
rate=soup7.find_all('span',class_="a-price-whole")
for i in rate:
    rating.append(i.text.replace('\n',''))
```

```python
import pandas as pd
amazon=pd.DataFrame({})
amazon['product']=product_name
amazon['Bowlers']=price
amazon['Nation']=image_url
amazon['Rating']=rating
```

Write a python program to extract information about the local weather from the National Weather Service
website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day
extended forecast display for the city. The data should include period, short description, temperature and
description.

```python
page8=requests.get("https://forecast.weather.gov/MapClick.php?textField1=37.78&textField2=-122.42#.YVlXTppBy00")
page8
soup8=BeautifulSoup(page8.content)
```

```python
heading=soup8.find("div",class_="panel-heading")
heading.text.replace('\n','  ')
```

```python
periods=[]
description=[]

period=soup8.find_all('div',class_="col-sm-2 forecast-label")
for i in period:
    periods.append(i.text.replace('\n',''))
    
des=soup8.find_all('div',class_="col-sm-10 forecast-text")
for i in des:
    description.append(i.text.replace('\n',''))
    
    
```

```python
import pandas as pd
weather=pd.DataFrame({})
weather['Period']=periods
weather['Description']=description
```
